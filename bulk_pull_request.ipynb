```python
from glob import glob
import os
import pandas as pd
import subprocess
from subprocess import PIPE
from tqdm import tqdm
import shutil
from time import sleep
```

```python
# csvs1 = glob("./menus_list_1/*.csv")
csvs4 = glob("./menus_list_4/*.csv")
csvs5 = glob("./menus_list_5/*.csv")
csvs = csvs4+csvs5
csvs = sorted(csvs, key=os.path.getsize)[::-1]
csvs = [os.path.abspath(file) for file in csvs]
branches = [os.path.splitext(os.path.basename(file))[0] for file in csvs]
```

```python
excluded_stores = ['walgreens', 'airshop', 'convenient-food-mart', 'liquor', 'convenience', 'food-mart', 'mart']
```

```python
from helpers import make_dolthub_pull_request
```

```python
from helpers import create_and_push_new_branch
```

```python
for branch, csv in tqdm(list(zip(branches, csvs))):
    if any(word in branch for word in excluded_stores):
        continue
    df = pd.read_csv(csv)
    restaurant_name = df['restaurant_name'].iloc[0].title()
    restaurant_name = restaurant_name.replace("'S", "'s")
    create_and_push_new_branch("../menus/", branch, csv, f"{restaurant_name} price data")
    sleep(30)
    description = "Scraped from UberEATS."
    response = make_dolthub_pull_request(restaurant_name, branch, description = description)
    if response.json().get('errors'):
        if response.json().get('errors')[0].get('message') == "fromBranch not found":
            continue
    shutil.move(csv, "./submitted/" + branch + ".csv")
```
